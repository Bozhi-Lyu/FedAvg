```python
import matplotlib.pyplot as plt
import statistics
from readData import *

overall_acc = results["oacc_i1"]
accs_bins_of_rounds = results["accs_bins_i1"]
predprob_of_rounds = results["pp_i1"]
ece_of_rounds = results["ece_i1"]
entropys_of_rounds = results["entro_i1"]
entrophys_bins_of_rounds = results["entro_bins_i1"]
VR_of_rounds = results["VR_i1"]
```

```python
assert len(predprob_of_rounds) == len(accs_bins_of_rounds), "Inconsistent list length"


x = []
y = []

for i in range(len(predprob_of_rounds)):
    assert len(predprob_of_rounds[i]) == len(accs_bins_of_rounds[i]), "Inconsistent inner list length"
    y.extend(predprob_of_rounds[i])
    x.extend(accs_bins_of_rounds[i])

plt.scatter(x, y, s = 1)

plt.plot([min(x), max(x)], [min(y), max(y)], color='black', linestyle='--')

plt.ylabel("Predicted Probability")
plt.xlabel("Accuracy")
plt.title("IID - Predicted Probability VS Accuracy")

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(range(1, len(ece_of_rounds) + 1), ece_of_rounds, color='blue', label='ECE Score')
ax.set_xlabel('Rounds')
ax.set_ylabel('ECE Score')
    
ax.yaxis.label.set_color('blue')
ax.tick_params(axis='y', colors='blue')
ax.spines['left'].set_color('blue')
ax.spines['right'].set_color('red')

ax2 = ax.twinx()
ax2.plot(range(1, len(overall_acc) + 1), overall_acc, marker='.', linestyle='-', color='red', label='Accuracy')
ax2.set_ylabel('Accuracy')
ax2.yaxis.label.set_color('red')
ax2.tick_params(axis='y', colors='red')
ax2.spines['right'].set_color('red')
ax2.spines['left'].set_color('blue')

fig.suptitle("IID - ECE, Acc VS Rounds")
fig.legend(loc='center right')

plt.show()
```
